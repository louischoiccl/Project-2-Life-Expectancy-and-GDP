```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import math
```

```python
gdp_df = pd.read_csv('all_data.csv')
gdp_df.rename(columns={'Life expectancy at birth (years)': 'LEABY'},inplace = True)
print(gdp_df.head())

```

```python
#Questions: Has life expectancy increased over time in the six nations?               
#Answer: Yes, the life expectancy increased over time in the six nations in long-term.
#Plot line plot of expectancy for the six nations, each nation one plot.
f = plt.subplots(2,3,figsize=(20,15))
#Chile 
ax_Chile = plt.subplot(231)
sns.lineplot(data = gdp_df, y = gdp_df.LEABY[gdp_df.Country == 'Chile'], x = 'Year',marker='o')
plt.title("Chile's LEABY")
plt.ylabel('Life expectancy at birth (years)')
ax_Chile.set_xticks(range(2000,2016))
plt.xticks(rotation=45)
#China 
ax_China = plt.subplot(232)
sns.lineplot(data = gdp_df, y = gdp_df.LEABY[gdp_df.Country == 'China'], x = 'Year',marker='o')
plt.title("China's LEABY")
plt.ylabel('Life expectancy at birth (years)')
ax_China.set_xticks(range(2000,2016))
plt.xticks(rotation=45)
#Germany
ax_Germany = plt.subplot(233)
sns.lineplot(data = gdp_df, y = gdp_df.LEABY[gdp_df.Country == 'Germany'], x = 'Year',marker='o')
plt.title("Germany's LEABY")
plt.ylabel('Life expectancy at birth (years)')
ax_Germany.set_xticks(range(2000,2016))
plt.xticks(rotation=45)
#Mexico
ax_Mexico = plt.subplot(234)
sns.lineplot(data = gdp_df, y = gdp_df.LEABY[gdp_df.Country == 'Mexico'], x = 'Year',marker='o')
plt.title("Mexico's LEABY")
plt.ylabel('Life expectancy at birth (years)')
ax_Mexico.set_xticks(range(2000,2016))
plt.xticks(rotation=45)
#United States of America
ax_USA = plt.subplot(235)
sns.lineplot(data = gdp_df, y = gdp_df.LEABY[gdp_df.Country == 'United States of America'], x = 'Year',marker='o')
plt.title("United States of America's LEABY")
plt.ylabel('Life expectancy at birth (years)')
ax_USA.set_xticks(range(2000,2016))
plt.xticks(rotation=45)
#Zimbabwe
ax_Zim = plt.subplot(236)
sns.lineplot(data = gdp_df, y = gdp_df.LEABY[gdp_df.Country == 'Zimbabwe'], x = 'Year',marker='o')
plt.title("Zimbabwe's LEABY")
plt.ylabel('Life expectancy at birth (years)')
ax_Zim.set_xticks(range(2000,2016))
plt.xticks(rotation=45)

plt.show()
plt.clf()

```

```python
#Question: Has GDP increased over time in the six nations?
#Answer: Yes GDP increased over time in the six nations of long-term run.
#Plot Line Plot of GDP for the six nations,each nation one plot
f = plt.subplots(2,3,figsize=(20,15))
#Chile 
ax_Chile = plt.subplot(231)
sns.lineplot(data = gdp_df, y = gdp_df.GDP[gdp_df.Country == 'Chile'], x = 'Year',marker='o')
plt.title("Chile's GDP")
plt.ylabel('GDP')
ax_Chile.set_xticks(range(2000,2016))
plt.xticks(rotation=45)
#China 
ax_China = plt.subplot(232)
sns.lineplot(data = gdp_df, y = gdp_df.GDP[gdp_df.Country == 'China'], x = 'Year',marker='o')
plt.title("China's GDP")
plt.ylabel('Life expectancy at birth (years)')
ax_China.set_xticks(range(2000,2016))
plt.xticks(rotation=45)
#Germany
ax_Germany = plt.subplot(233)
sns.lineplot(data = gdp_df, y = gdp_df.GDP[gdp_df.Country == 'Germany'], x = 'Year',marker='o')
plt.title("Germany's GDP")
plt.ylabel('GDP')
ax_Germany.set_xticks(range(2000,2016))
plt.xticks(rotation=45)
#Mexico
ax_Mexico = plt.subplot(234)
sns.lineplot(data = gdp_df, y = gdp_df.GDP[gdp_df.Country == 'Mexico'], x = 'Year',marker='o')
plt.title("Mexico's GDP")
plt.ylabel('GDP')
ax_Mexico.set_xticks(range(2000,2016))
plt.xticks(rotation=45)
#United States of America
ax_USA = plt.subplot(235)
sns.lineplot(data = gdp_df, y = gdp_df.GDP[gdp_df.Country == 'United States of America'], x = 'Year',marker='o')
plt.title("United States of America's GDP")
plt.ylabel('GDP')
ax_USA.set_xticks(range(2000,2016))
plt.xticks(rotation=45)
#Zimbabwe
ax_Zim = plt.subplot(236)
sns.lineplot(data = gdp_df, y = gdp_df.GDP[gdp_df.Country == 'Zimbabwe'], x = 'Year',marker='o')
plt.title("Zimbabwe's GDP")
plt.ylabel('GDP')
ax_Zim.set_xticks(range(2000,2016))
plt.xticks(rotation=45)

plt.show()
plt.clf()
```
